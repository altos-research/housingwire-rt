```python
import pandas as pd
```

# 2023 Data

```python
# df = pd.read_csv('./raw/2024_submissions_static.csv')
df = pd.read_csv('./raw/2024_submissionsView_Will.csv')
df.rename(columns={'Title Policies & Services Volulme':'Title Policies & Services Volume',
                    '2024 avg home price':'Avg Home Price'}, inplace=True)
df.info()
```

## 2023 Mortgages

```python
mortgage = df.loc[~df['Mortgage Loan Transactions'].isnull(), ['CompanyID', 'Company', 'Mortgage Loan Transactions', 'Mortgage Loan Volume']].reset_index(drop=True)
# clean this: "30.35M (total transactions overall)"
mortgage.loc[mortgage['Mortgage Loan Volume']=='30.35M (total transactions overall)', 'Mortgage Loan Volume'] = str(30.35 * 1000000)
# clean punctuation
mortgage['Mortgage Loan Volume'] = [int(x.replace(',','').replace('.','').replace('$','')) for x in mortgage['Mortgage Loan Volume']]
mortgage['Mortgage Loan Transactions'] = [int(x.replace(',','').replace('.','').replace('$','')) for x in mortgage['Mortgage Loan Transactions']]
mortgage
```

```python
mortgage.to_excel('./cleaned/MortgageCos.xlsx', index=False)
```

## 2023 Title

```python
titles = df.loc[~df['Title Policies & Services Transactions'].isnull(), ['CompanyID', 'Company', 'Title Policies & Services Transactions', 'Title Policies & Services Volume']].reset_index(drop=True)
titles['Title Policies & Services Volume'] = [float(str(x).replace('$','').replace(',','')) if x != '' else x for x in titles['Title Policies & Services Volume']]
titles['Title Policies & Services Transactions'] = titles['Title Policies & Services Transactions'].str.replace(',','').astype('int')
titles
```

```python
titles.to_excel('./cleaned/TitleCos.xlsx', index=False)
```

## 2023 Home Warranty

```python
warranty = df.loc[~df['Home Warranty Policies Sold'].isnull(), ['CompanyID', 'Company', 'Home Warranty Policies Sold']].reset_index(drop=True)
# drop 'none' value
warranty = warranty.loc[warranty['Home Warranty Policies Sold']!='none']
warranty['Home Warranty Policies Sold'] = warranty['Home Warranty Policies Sold'].astype('int')
warranty.sort_values(by='Home Warranty Policies Sold', ascending=False, inplace=True)
warranty.reset_index(drop=True, inplace=True)
warranty
```

```python
warranty.to_excel('./cleaned/WarrantyCos.xlsx', index=False)
```

## 2023 Escrow

```python
escrow = df.loc[~df['Escrow Policies Sold'].isnull(), ['CompanyID', 'Company', 'Escrow Policies Sold']].reset_index(drop=True)
# drop Pamela
escrow = escrow.loc[escrow['Escrow Policies Sold']!='Pamela']
# drop nulls
escrow['Escrow Policies Sold'] = escrow['Escrow Policies Sold'].str.replace(',','').astype('int')
escrow = escrow.loc[escrow['Escrow Policies Sold']>0].sort_values(by='Escrow Policies Sold', ascending=False).reset_index(drop=True)
escrow
```

```python
escrow.to_excel('./cleaned/EscrowCos.xlsx', index=False)
```

## 2023 Brokers

```python
df.drop(columns=[
    'Mortgage Loan Transactions', 'Mortgage Loan Volume',
    'Title Policies & Services Transactions', 'Title Policies & Services Volume',
    'Home Warranty Policies Sold',
    'Escrow Policies Sold',
    'Homeowners Insurance Policies Sold'
], inplace=True)
```
